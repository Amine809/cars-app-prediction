##Converting categorical data to numerical format

```python
def convtransmission(x):
  if x=='Automatic':
    return 1
  elif x=='Manual':
    return 0

  else:
    return -1


new_cars_end['transmission']=new_cars_end['transmission'].map(convtransmission)
```

```python
def convfuel(x):
  if x=='Gasoline':
    return 1
  elif x=='Hybrid':
    return 0
  elif x=='E85 Flex Fuel':
    return 2
  elif x=='Diesel':
    return -1


  else:
    return -2


new_cars_end['fuel_type']=new_cars_end['fuel_type'].map(convfuel)
```

```python
def convclean(x):
  if x=='Yes':
    return 1
new_cars_end['clean_title']=new_cars_end['clean_title'].map(convclean)
```

##Beginning with machine learning model implementation

```python
# Target variable (y)
y = new_cars_end['price']
X=new_cars_end.drop(['price','brand','model','ext_col','int_col'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.decomposition import PCA
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
Xtrain_scaled, Xtest_scaled, ytrain, ytest = train_test_split(X_scaled, y, test_size=0.15, random_state=42)
```

```python
# Hyperparameter Tuning
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
random_search = RandomizedSearchCV(RandomForestRegressor(random_state=42, n_jobs=-1), param_distributions=param_grid, n_iter=10, cv=10, verbose=2, random_state=42, n_jobs=-1)
random_search.fit(Xtrain_scaled, ytrain)
```

```python
# Get the best parameters
best_params = random_search.best_params_
print("Best Parameters:", best_params)
```

```python
# Apply PCA if desired
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)
Xtrain_pca, Xtest_pca, ytrain, ytest = train_test_split(X_pca, y, test_size=0.15, random_state=42)
```

```python
# Train the model with the best parameters
final_model = RandomForestRegressor(**best_params, random_state=42, n_jobs=-1)
final_model.fit(Xtrain_scaled, ytrain)
```

```python
# Evaluate the model
final_model_accuracy_train = round(final_model.score(Xtrain_scaled, ytrain) * 100, 2)
print("Training Accuracy:", final_model_accuracy_train, "%")
```

```python
final_model_accuracy_test = round(final_model.score(Xtest_scaled, ytest) * 100, 2)
print("Testing Accuracy:", final_model_accuracy_test, "%")
```

```python
# Make predictions for a new observation
new_observation = pd.DataFrame({
    'model_year': [2018	],
    'fuel_type': [1],
    'transmission': [1],
    'clean_title': [1],
    'mil': [53705],
    'accidents': [1],
    'horsepower': [241.0],
    'motor(l)': [2.0],
    'number_of_cylinders': [4],
})

# Scale the new observation
new_observation_scaled = scaler.transform(new_observation)

# Make a prediction for the new observation
prediction = final_model.predict(new_observation_scaled)

# Print the predicted price
print('Predicted Price:', prediction)
```
